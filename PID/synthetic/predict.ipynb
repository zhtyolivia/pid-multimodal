```python
import numpy
import sys
import os
import pickle
import torch
sys.path.append('/usr0/home/yuncheng/MultiBench/')
from supervised_learning import test
from ensemble import test as test_ensemble
from get_data import get_dataloader
from rus import *
```

```python
if os.path.isfile('synthetic/experiments/results.pickle'):
    with open('synthetic/experiments/results.pickle', 'rb') as f:
        results = pickle.load(f)
else:
    results = dict()
for method in ['additive', 'agree', 'align', 'early_fusion', 'elem', 'mfm', 'mi', 'mult', 'outer', 'lower']:
    if method not in results:
        results[method] = dict()
    for SETTING in ['redundancy', 'uniqueness0', 'uniqueness1', 'synergy', 'mix1', 'mix2', 'mix3']:
        print(SETTING)
        data_path = 'synthetic/DATA_{}.pickle'.format(SETTING)
        saved_model = 'synthetic/experiments/{}/{}_{}_best.pt'.format(SETTING, SETTING, method)
        saved_cluster = 'synthetic/experiments/{}/{}_{}_cluster.pickle'.format(SETTING, SETTING, method)
        _, _, _, testdata = get_dataloader(path=data_path, keys=['0','1','label'], modalities=[1,1], batch_size=128, num_workers=4)
        model = torch.load(saved_model).cuda()
        num_params = sum([p.numel() for p in model.parameters()])
        if method in ['additive', 'agree', 'align']:
            acc = test_ensemble(model, testdata, no_robust=True, criterion=torch.nn.CrossEntropyLoss(), save_preds=saved_cluster)
        else:
            acc = test(model, testdata, no_robust=True, criterion=torch.nn.CrossEntropyLoss(), save_preds=saved_cluster)
        with open(saved_cluster, 'rb') as f:
            preds = pickle.load(f)
        with open('synthetic/DATA_{}_cluster.pickle'.format(SETTING), 'rb') as f:
            cluster = pickle.load(f)
        pred_results = (cluster['test']['0'], cluster['test']['1'], preds.reshape(-1,1))
        P, maps = convert_data_to_distribution(*pred_results)
        tmp = get_measure(P)
        if SETTING not in results[method]:
            results[method] = tmp
        else:
            for k in tmp:
                results[method][SETTING][k] = tmp[k]
        results[method][SETTING]['acc'] = acc
        results[method][SETTING]['params'] = num_params
        print()

with open('synthetic/experiments/results.pickle', 'wb') as f:
    pickle.dump(results, f)
```

```python
if os.path.isfile('synthetic/experiments/results.pickle'):
    with open('synthetic/experiments/results.pickle', 'rb') as f:
        results = pickle.load(f)
else:
    results = dict()
for method in ['additive', 'agree', 'align', 'early_fusion', 'elem', 'mfm', 'mi', 'mult', 'outer']:
# for method in ['tf']:
    print(method)
    for SETTING in ['redundancy', 'uniqueness0', 'uniqueness1', 'synergy', 'mix1', 'mix2', 'mix3', 'mix4', 'mix5', 'mix6']:
        print(SETTING)
        for i, modality in enumerate([[1,0], [0,1]]):
            print('unimodal', i)
            data_path = 'synthetic/DATA_{}.pickle'.format(SETTING)
            saved_model = 'synthetic/experiments/{}/{}_{}_best.pt'.format(SETTING, SETTING, method)
            _, _, _, testdata = get_dataloader(path=data_path, keys=['0','1','label'], modalities=modality, batch_size=128, num_workers=4)
            model = torch.load(saved_model).cuda()
            if method in ['additive', 'agree', 'align']:
                acc = test_ensemble(model, testdata, no_robust=True, criterion=torch.nn.CrossEntropyLoss())
            else:
                acc = test(model, testdata, no_robust=True, criterion=torch.nn.CrossEntropyLoss())
            results[method][SETTING]['unimodal{}_acc'.format(i)] = acc
        print()

with open('synthetic/experiments/results.pickle', 'wb') as f:
    pickle.dump(results, f)
```
